```python
import pandas as pd
import unicodedata
from datetime import date, timedelta, datetime
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.float_format = None
# pd.options.display.float_format = '{:.0f}'.format
```

```python
dfo = pd.read_csv('output/spk_daily_over.csv')
dfo.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
dfo.set_index("date", inplace=True)
dfo = dfo[(dfo['over_diff'] <= .202) & (dfo['over_odds'] <= 0.61)]

dfu = pd.read_csv('output/spk_daily_under.csv')
dfu.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
dfu.set_index("date", inplace=True)
dfu = dfu[(dfu['under_diff'] <= .202) & (dfu['under_odds'] <= 0.61)]
```

```python
def float_range(start, end, increment):
    current_value = start
    while current_value <= end:
        yield current_value
        current_value += increment

start = 0.04
end = 0.17
increment = 0.01

my_range_generator = list(float_range(start, end, increment))
odds_range = list(float_range(.42, .55, .01))
```

```python
# results_df = pd.DataFrame()
result_columns = ['odds', 'value', 'dfo_total_earnings', 'dfo_total_wins', 'dfo_total_losses', 'dfo_win_ratio']
results_df = pd.DataFrame(columns=result_columns)

for odds in odds_range:
    for value in my_range_generator:
        # print(odds)
        dfo_copy = dfo.copy()  # Make a copy of dfo DataFrame
        dfo_copy = dfo_copy[(dfo_copy['over_odds'] >= odds) & (dfo_copy['over_diff'] >= value)]
        dfo_total_earnings = dfo_copy['wager_return'].sum()
        dfo_total_wins = len(dfo_copy[dfo_copy['win'] == True])
        dfo_total_losses = len(dfo_copy[dfo_copy['win'] == False])

        if dfo_total_wins + dfo_total_losses == 0:
            dfo_win_ratio = 0
        else:
            dfo_win_ratio = round(dfo_total_wins / (dfo_total_wins + dfo_total_losses), 3)

        temp_row = [odds, round(value, 2), 
                    round(dfo_total_earnings, 2), dfo_total_wins, dfo_total_losses, dfo_win_ratio]
        
        results_df.loc[len(results_df)] = temp_row

# Display the resulting DataFrame
# print(results_df)
```
